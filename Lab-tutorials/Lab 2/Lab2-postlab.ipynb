```python
import matplotlib.pyplot as plt
import random

start = int(input("Choose a start year"))
end = int(input("Choose an end year"))
cm = plt.get_cmap('YlGn')

NUM_COLORS = end-start+1
plot = plt.figure()
ax = plot.add_subplot(111)
ax.set_prop_cycle(color = [cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)])

months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

plt.title("Temperature Yea!!!", fontdict = {'fontname': 'Comic Sans MS', 'fontsize':20})
plt.xlabel('Months')
plt.ylabel('Temperature Â°C')

for year in range(start, end+1):
    year_temp = []
    for month in range(0,12):
        month_temp = int(input("Please enter the average temperature in "+str(months[month]+" of "+ str(year))))
        year_temp.append(month_temp)
    ax.plot(months, year_temp, label = str(year), marker = 'o', markersize=6)
    ax.legend(loc = "upper left")
    plt.tight_layout()
plt.show()
```
